```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Sample data
warehouses = ['Warehouse A', 'Warehouse B', 'Warehouse C']
stock_levels = [100, 150, 80]
demand = [90, 120, 100]
orders_over_time = {
    'Warehouse A': [5, 10, 15, 20, 25],
    'Warehouse B': [10, 15, 20, 25, 30],
    'Warehouse C': [15, 20, 25, 30, 35]
}
time = ['Jan', 'Feb', 'Mar', 'Apr', 'May']

# Bar chart for stock levels and demand
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
bar_width = 0.35
index = np.arange(len(warehouses))

ax[0].bar(index, stock_levels, bar_width, label='Stock Levels')
ax[0].bar(index + bar_width, demand, bar_width, label='Demand')

ax[0].set_xlabel('Warehouses')
ax[0].set_ylabel('Quantity')
ax[0].set_title('Stock Levels and Demand')
ax[0].set_xticks(index + bar_width / 2)
ax[0].set_xticklabels(warehouses)
ax[0].legend()

# Line chart for orders over time
for warehouse, orders in orders_over_time.items():
    ax[1].plot(time, orders, label=warehouse)

ax[1].set_xlabel('Time')
ax[1].set_ylabel('Orders')
ax[1].set_title('Orders Over Time')
ax[1].legend()

plt.tight_layout()
plt.show()

# Heatmap for stock-demand comparison
comparison = np.array(stock_levels) - np.array(demand)
heatmap_data = pd.DataFrame(comparison.reshape(1, -1), columns=warehouses)

plt.figure(figsize=(6, 2))
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', cbar=False)
plt.title('Stock vs Demand Comparison')
plt.show()
```
